```python
from sklearn import tree
```

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import pydotplus
from IPython.display import Image
```

```python
# Generate synthetic data
X, y = make_classification(n_samples=1500, n_features=2, n_informative=2, n_redundant=0, n_repeated=0,
n_classes=2, n_clusters_per_class=1, random_state=42)
```

```python
clf=tree.DecisionTreeClassifier(criterion='entropy',max_depth=3)
```

```python
clf
```

```python
#Training the DT
clf=clf.fit(X,y)
```

```python
clf
```

```python
#Exporting the DT to Graphviz file
dot_data = export_graphviz(clf, out_file=None,
feature_names=['Feature 1', 'Feature 2'],
class_names=['Class 0', 'Class 1'],
filled=True, rounded=True,
special_characters=True)
```

```python
graph = pydotplus.graph_from_dot_data(dot_data)
```

```python
Image(graph.create_png())
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2, random_state=11)
```

```python
x_test.shape
```

```python
from sklearn import tree
```

```python
from sklearn.metrics import accuracy_score
```

```python
maxDepth=[2,3,4,5,6,7,8,9,10,20,25,30,35,40,45,50,55]
```

```python
trainAccuracy=np.zeros(len(maxDepth))
```

```python
print(trainAccuracy)
```

```python
testAccuracy=np.zeros(len(maxDepth))
```

```python
print(testAccuracy)
```

```python
index = 0
```

```python

from sklearn.metrics import accuracy_score
```

```python
for depth in maxDepth:
  clf=tree.DecisionTreeClassifier(max_depth=depth)
  clf=clf.fit(x_train,y_train)
  y_predictTrain=clf.predict(x_train)
  y_predictTest=clf.predict(x_test)
  trainAccuracy[index]=accuracy_score(y_train,y_predictTrain)
  testAccuracy[index]=accuracy_score(y_test,y_predictTest)
  index +=1
```

```python
trainAccuracy
```

```python
testAccuracy
```

```python
#matplotlib inline
plt.plot(maxDepth,trainAccuracy,'ro-',maxDepth, testAccuracy, 'bv--')
plt.legend(['trainAccuracy','testAccuracy'])
plt.xlabel('maxDepth')
plt.ylabel('Accuracy')
```
